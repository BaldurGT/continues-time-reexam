> Opgave 5. **Foldning.**

Udgangssignalet y(t) fra et LTIC system er givet ved foldningen:
$$y(t)= (3e^{-3t}u(t))*(5te^{-5t}u(t))$$
hvor * er foldningsoperatoren og u(t) er enhedstrinpåvirkningen (også kaldet enhedsstepfunktionen). $\\$
Tidsvariablen t har enheden sekund (s). $\\$
Koefficienters og konstante leds enheder er ikke medtaget men skal antages at være korrekte. $\\$
Afkryds det korrekte udtryk for y(t). Benyt eventuelt tabelopslag. $\\$


```python
from sympy import *
t, s = symbols('t s')

f1_t = 3 * exp(-3*t) * Heaviside(t)
f2_t = 5 * t * exp(-5*t) * Heaviside(t)

# Laplace transformation:
F1_s, _, _ = laplace_transform(f1_t, t, s)
F2_s, _, _ = laplace_transform(f2_t, t, s)

# Foldning i s-domænet (multiplikation)
Y_s = F1_s * F2_s
display(Y_s)

# Eventuelt delbrøkopdeling for lettere invers
Y_s_apart = apart(Y_s)
display(Y_s_apart)

# Invers Laplace transform for at finde y(t)
y_t = inverse_laplace_transform(Y_s_apart, s, t)
display(y_t)
```
