> Opgave 6: $\textbf{Fourierrækker.}$ (4)

Om Fourierrække-ekspansion af et signal gives følgende sande eller falske udsagn:

$\textbf{Udsagn A.} \\$
Lad $f(nw_0t)$ være en basisfunktion i en Fourierrække-ekspansion. Det indre produkt mellem to basisfunktioner er kun forskellig fra nul, hvis der er tale om det indre produkt af en basisfunktion med sig selv.

$\textbf{Udsagn B.} \\$
Ekspanderes et tidsudsnit $t_1 \leq t \leq t_1 + T_0$ af et reelt og aperiodisk signal $z(t)$ i en Fourierrække, vil resultatet af Fourierrække-ekspansion $\hat{x}(t)$ være reelt og periodisk med periodetiden $T_0$

$\textbf{Udsagn C.} \\$
Om koefficienterne $D_n$ i den kompleks-eksponentielle Fourierrække-ekspansion af et reelt og ulige signal $x(t)$ gælder: $|D_n|=-|D_{-n}|$

$\textbf{Udsagn D.} \\$
Om koefficienterne $D_n$ i den kompleks-eksponentielle Fourierrække-ekspansion af et reelt og lige signal $x(t)$ gælder: $D_n = D_{-n}$

> Vælg én svarmulighed:
* Udsagn A og B er sande.
* Udsagn B og C er sande.
* Udsagn A og D er sande.
* Udsagn C og D er sande.
* Ingen af de øvrige svarmuligheder indeholder kun sande udsagn.
* Der er mere end én svarmulighed, der udelukkende indeholder sande udsagn.

> Opgave 19: $\textbf{Instrumenteringsforstærker.}$

$R_{1a}=R_{1b}=R_{\frac{G}{2}}$, hvor $R_G$ er den eksterne modstand til justering af forstærknig. $\\$
$R_{2a}=R_{2b}; \space R_{3a}=R_{3b}=R_{4a}=R_{4b}$

> Opgave 20 $\textbf{Sampling og analog-til-digital konvertering.}$

$\textbf{Udsagn A.} \\$
Et signal med Fourier transformationsparret $x(t) \leftrightarrow X(w)$ samples med tidsintervallet $\textbf{T}$ mellem samples.$\\$
Det samplede signal $\hat{x}(t) = x(t) \sum^\infty_{n=-\infty} \delta(t-nT)$ har et frekvensspektrum beskrevet ved:
$$\hat{X}(w)=\frac{1}{T}\sum^\infty_{n=-\infty} X(w-nw_s)$$
Det følger at frekvensspektret for det samplede signal er diskret med en afstand $w$, mellem frekvenskomposanterne på frekvensaksen.
