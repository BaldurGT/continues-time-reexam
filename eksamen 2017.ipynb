> Opgave 13

**Et 2. ordens LTIC system har diagrammet i figuren:** $\\$
 $\\$
**Hvad er systemets overføringsfunktion i laplacedomænet?** $\\$


**B er den korekte**



**Output Y(s) kan defineres som:**
$$Y(s)=10 \cdot e(s)$$

**Dette kan udregnes som**
$$z_1(s)=\frac{e(s)}{s}$$
$$z_2(s)=\frac{z_1(s)}{s}=\frac{e(s)}{s^2}$$
$$sum2(s)=6\cdot z_1(s)+25z_2(s) = \frac{6e(s)}{s}+\frac{25e(s)}{s^2}$$
$$e(s)=X(s)-sum2(s)=X(s)-(\frac{6e(s)}{s}+\frac{25e(s)}{s^2})$$

**Multiplicer begge sider med $s^2$ for at fjerne nævnere:**
$$s^2e(s)+6se(s)+25e(s)=s^2X(s) \Rightarrow e(s)(s^2+6s+25)=s^2X(s) \Rightarrow e(s)=\frac{s^2}{s^2+6s+25X}X(s)$$

**Nu finder jeg H:** 
$$Y(s)=10 \cdot e(s)= 10 \cdot \frac{s^2}{s^2+6s+25X}X(s) \Rightarrow H(s) = \frac{Y(s)}{X(s)}= \frac{10s^2}{s^2+6s+25}$$
