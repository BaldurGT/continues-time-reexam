$\textbf{Jeg kan finde dempningen med at sammenligne systemets differentialligning} \\ \textbf{med standardformen for et 2. ordens system:} \\$

| $\zeta$ | Betegnelse | Systemrespons |
| ------- | ---------- | ------------- |
| $\mathbf{0<\zeta < 1}$ | **Underdæmpet** | Oscillerer (ringer) med aftagende amplitude. Jo mindre $\zeta$, jo mere svinger det. (overshoot)|
| $\mathbf{\zeta > 1}$   | **Overkritisk dæmpet** | Ingen oscillation, men langsommere tilbagevenden end kritisk dæmpet. (intet overshoot)|
| $\mathbf{\zeta = 1}$   | **Kritisk dæmpet** | Ingen oversving, vender hurtigst muligt tilbage til ligevægt uden oscillation. (intet overshoot)|
| $\mathbf{\zeta = 0}$   | **Udæmpet** | Ingen oscillation, men langsommere tilbagevenden end kritisk dæmpet. (maximal oscillation)|

> 3. Jeg finder enhets impus responsen:

$\textbf{Jeg begynder med at tage lapace af begge sider} \\$
$$\ddot{y}(t)+8\dot{y}(t) + 16y(t) = 32\dot{x}(t)$$

$\mathcal{L}\{ \ddot{y}(t) \} = s^2Y(s)-sy(0_-)-\dot{y}(0_-) \\ $
$\mathcal{L}\{ \dot{y}(t)  \} = sY(s)-y(0_-) \\ $ 
$\mathcal{L}\{ y(t)        \} = Y(s) \\ $

$\ddot{y}(t)=s^2Y(s)$,   $\quad 8\dot{y}(t)=8sY(s)$,   $\quad 16y(t)=16Y(s)$,   $\quad 32\dot{x}(t)=32sX(s)$

$s^2Y(s) + 8sY(s) + 16Y(s) = 32sX(s) \Rightarrow (s^2+8s+16)Y(s) = 32sX(s)$

$\textbf{Nu kan jeg finde overførings funktionen: } H(s)=\frac{Y(s)}{X(s)} \\$
$H(s)=\frac{32s}{s^2+8s+16}$

Fordi: $ s^2+8s+16 = (s+4)^2 : \\$
$H(s)=\frac{32s}{(s+4)^2}$

$\textbf{Nu udfører vi delbrøksopdeling:} \\$
Når nævneren har et gentaget førstegradspolynomium $(s+4)^2$, skriver vi: $\\$
$\frac{32s}{(s+4)^2} = \frac{A}{(s+4)} + \frac{B}{(s+4)^2} \\$
hvor A og B er konstanter, vi skal finde.

Nu finder jeg fælles nævner ved at gange begge sider med $(s+4)^2: \\$
$\frac{32s}{(s+4)^2} = \frac{A}{(s+4)} + \frac{B}{(s+4)^2} \Rightarrow 32s = A(s+4)+B$

Nu kan jeg finde konstanten B ved at sette s=-4, og bruger verdien for B, samt s=0 for at finde A: $\\$
$s=-4:\\$
$32s = A(s+4)+B \Rightarrow 32(-4) = A(-4+4)+B \Rightarrow -128 = A(0)+B \Rightarrow B=-128$

$s=0:\\$
$32s = A(s+4)+B \Rightarrow 32(0) = A(0+4)-128 \Rightarrow A = \frac{128}{4} \Rightarrow A = 32$

Resultat: $\\$
$
\frac{32s}{(s+4)^2} = \frac{A}{(s+4)} + \frac{B}{(s+4)^2} \Rightarrow 
\frac{32s}{(s+4)^2} = \frac{32}{(s+4)} - \frac{128}{(s+4)^2} \Rightarrow
\frac{32s}{(s+4)^2} = 32(\frac{1}{(s+4)} - \frac{4}{(s+4)^2})
$

$\textbf{Nu kan jeg bruge den invers-Laplace for at finde enhed-impulsresponsen h(t)}:$
$\frac{32s}{(s+4)^2} = 32(\frac{1}{(s+4)} - \frac{4}{(s+4)^2})$

$\mathcal{L}^-1\{\frac{1}{s+a}\}=e^{-at}u(t)$, $\mathcal{L}^-1\{\frac{1}{(s+a)^2}\}=te^{at}u(t) \\$

Her er $a=4 \\$
$
h(t) = \mathcal{L}^-1\{H(s)\} = 32[\mathcal{L}^-1\{\frac{1}{s+4}\} - \mathcal{L}^-1\{\frac{1}{(s+4)^2}\}] \Rightarrow
32[e^{-4t}-4te^{-4t}]u(t) \Rightarrow 32(1-4t)e^{-4t}u(t)
$

$\textbf{Enhed-impulsresponsen h(t) er: } 32(1-4t)e^{-4t}u(t)$

```python
s, t = symbols('s t')

# Define the Laplace-domain impulse response (ie H(s))
H = (0 + 32*s) / (s**2 + 8*s + 16)  # 5*s^2 + 0*s^1 + 0  / ( s^2 + 5*s^1 + 5*s^0 )
h = inverse_laplace_transform(H, s, t)
simplify(h)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, step

# Define a time vector
t = np.linspace(0, 10, 1000)

# --- LOWPASS FILTER ---
# Second-order lowpass: H(s) = ω_n^2 / (s^2 + 2ζω_n s + ω_n^2)
omega_n = 1
zeta = 0.5  # underdamped
num_lp = [omega_n**2]
den_lp = [1, 2*zeta*omega_n, omega_n**2]
system_lp = TransferFunction(num_lp, den_lp)
t_lp, y_lp = step(system_lp, T=t)

# --- HIGHPASS FILTER ---
# Second-order highpass: H(s) = s^2 / (s^2 + 2ζω_n s + ω_n^2)
num_hp = [1, 0, 0]
den_hp = [1, 2*zeta*omega_n, omega_n**2]
system_hp = TransferFunction(num_hp, den_hp)
t_hp, y_hp = step(system_hp, T=t)

# --- BANDPASS FILTER ---
# Second-order bandpass: H(s) = 2ζω_n s / (s^2 + 2ζω_n s + ω_n^2)
num_bp = [0, 2*zeta*omega_n, 0]
den_bp = [1, 2*zeta*omega_n, omega_n**2]
system_bp = TransferFunction(num_bp, den_bp)
t_bp, y_bp = step(system_bp, T=t)

# Plot the responses
plt.figure(figsize=(10, 6))
plt.plot(t_lp, y_lp, label="Lowpass", linewidth=2)
plt.plot(t_hp, y_hp, label="Highpass", linewidth=2)
plt.plot(t_bp, y_bp, label="Bandpass", linewidth=2)
plt.title("Step Response of Second-Order Filters")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

$\textbf{Basis funktion:} \\$
En basis funktion er en funktion der kan bruges til at bygge op andre funktioner fx. det normale exemple er sinus og cosnus funtioner.

$\textbf{Fourierrække-ekspansion:} \\$
En fourierrække-ekspansion er en måtte at skriv en periodisk funktion op som en sum af basis funktioner. $\\$
Ideen er at ethvert “pænt” periodisk signal kan skrives som en kombination af simple svingninger.

Synteseformel (genskab signalet):
$$x(t)=\sum^\infty_{n=-\infty} D_n e^{jnw_0t}$$
$e^{jnω_0t}$ er basisfunktionerne, og $D_n$ er koefficienterne, der fortæller, hvor meget af hver basisfunktion der skal med.

* Basisfunktioner = “ingredienser”
* Koefficienter = “mængden af hver ingrediens”
* Summen = “det færdige måltid” (dit signal)

$\textbf{Fourierkoefficienter (Indre produkt):} $
Et indre produkt er et “generel prikprodukt” der måler, hvor meget to funktioner minder om hinanden. Vi bruger fromelen for Fourierrække men vi bruger Analyseformelen.$\\$

Analyseformel (find koefficienterne):
$$D_n=\frac{1}{T_0}\int^{t_0+T_0}_{t_0} x(t)e^{-jnw_0t}dt$$

Alternativt kan det skrives med indre-product notation, en mere kompakt måtte at skrive formelen for $D_n$ på:
$$⟨f,g⟩ = \frac{1}{T}\int_{t_0}^{yt_0+T}f(t)g^*(t)dt$$
Hvor $g^∗$ betyder “kompleks konjugat” (man ændrer fortegnet på den imaginære del af et komplekst tal.) — for rene sinus/cosinus kan du bare ignorere stjernen.

Hvis $⟨f,g⟩=0$, er de ortogonale → ingen “overlapning” mellem dem. $\\$
Hvis $⟨f,f⟩>0$, måler det “energien” i funktionen. $\\$
I Fourier-teorien er basisfunktionerne valgt, så de er ortogonale — dvs. deres indre produkt er nul, medmindre det er funktionen med sig selv.

$\textbf{Opsumering:} \\$
Basisfunktioner som “byggeklodser” $\\$
Indre produkt som “overlapningsmåling” $\\$
Fourierrække som “byggeklods-sum”. $\\$

> Opgave 1. Fourier-trasformation vs. Fourierrække-ekspansion.

$\textbf{Fourier-trasformation} \\$
En generalisering af Fourierrækken, der bruges til at repræsentere aperiodiske signaler som en sum/integral af uendeligt mange frekvenser. $\\$
Når perioden $T_0\rightarrow\infty$, bliver frekvenserne kontinuerte, og summen i Fourierrækken bliver til et integral.

Analyseformel:
$$X(w)=\int^\infty_{-\infty}x(t)e^{-jwt}dt$$
hvor $X(w)$ er det kontinuerte frekvensspektrum.

Synteseformel:
$$x(t)=\frac{1}{2\pi}\int^\infty_{\infty}X(w)e^{jwt}dw  $$

$\textbf{Fourierrække-ekspansion} \\$
En måde at repræsentere et periodisk signal x(t) som en sum af harmoniske bølger (sinus, cosinus eller komplekse eksponentialer). $\\$
Et periodisk signal kan bygges op som en “opskrift” af basisfunktioner $e^{jnω_0t}$ med forskellige vægte (Fourierkoefficienterne).

Fundamental vinkelhastighed:
$$w_0 = \frac{2\pi}{T_0}$$
$T_0$ er periodetiden.

Analyseformel (find koefficienterne $D_n$):
$$D_n=\frac{1}{T_0} \int^{t_0+T_0}_{t_0} x(t)e^{-jnw_0t}dt$$

Synteseformel (genskab signalet):
$$x(t)=\sum^\infty_{n=-\infty} D_ne^jnw_0t$$

$\textbf{Konklusjon:} \\$
Det er ikke rigtigt, at Fouriertransformationen “kun” eksisterer for aperiodiske signaler. $\\$
Den eksisterer også for periodiske signaler, men den ser anderledes ud (består af impulser i frekvensdomænet). $\\$
Det er korrekt, at vi normalt bruger Fourierrækken til periodiske signaler, fordi det er mere praktisk, men matematikken bag Fouriertransformationen virker stadig. $\\$

> Opgave 2. Reel og imaginær del af fourier transformation, af lige eller ulige signaler. 

$\textbf{Vi bruger definitionen af Fourier-transformen og paritet.} \\$
$$X(w)=\int^\infty_{-\infty}x(t)e^{-jwt}dt = \int^\infty_{-\infty}x(t)[cos(wt)-j\cdot sin(wt)]dt$$

Sæt:

$$\mathcal{R}\{X(w)\} = \int^\infty_{-\infty} x(t)cos(wt)dt, \qquad \mathcal{I}\{X(w)\} = -\int^\infty_{-\infty}x(t)sin(wt)dt$$
* cos(ωt) er lige; sin(ωt) er ulige.
* Produktregler: $lige·lige = lige, \quad ulige·ulige = lige, \quad lige·ulige = ulige, \quad ulige·lige = ulige.$
* Integral af en ulige funktion over $(-\infty, \infty)$ er 0.

$\textbf{Case 1:}$ $x(t)$ er reel og lige $\\$
* $x(t)cos(wt): \quad lige \cdot lige = lige \rightarrow$ kan være $\neq 0 \rightarrow$ giver realdel.
* $x(t)sin(wt): \quad lige \cdot ulige = ulige \rightarrow $ integral = 0 \rightarrow \mathcal{I}\{X\}=0$.

Fourier-transformen er ren reel (og lige). Påstanden 1 er sand.

$\textbf{Case 2:}$ $x(t)$ er reel og ulige $\\$
* $x(t)cos(wt): \quad ulige \cdot lige = ulige \rightarrow $ integral $=0\rightarrow\mathcal{R}\{X\}=0$.
* $x(t)cos(wt): \quad ulige \cdot ulige = lige \rightarrow $ kan være $\neq 0 \rightarrow$ giver imaginærdel.

Fourier-transformen er ren imaginær (og ulige). Påstanden 2 er sand.

Konklusjon: påstanden er sand.

$$H(jw) = \frac{Y(jw)}{X(jw)} = 1+\frac{Z_f}{Z_g} \Rightarrow H(jw) = 1+\frac{\frac{R_2}{1+jwR_2C_1}}{R_1} = 1+\frac{R_1}{R_2}\cdot\frac{1}{1+jwR_2C_1}$$

$\textbf{Hvad vis feedback var istede koblet til } V_+? \\$
Når feedback i stedet går til $v_+$ er det positiv feedback. Så kan vi ikke bruge “virtual short” $(v_−=v_+)$-tricket. Man må tage den generelle model med op-ampens åben-sløjfe forstærkning $A(jω)$ og feedbackfaktoren
$$B(jw) = \frac{Z_g}{Z_f+Z_g}$$

Her fra kan man få formlerne: $\\$
Negativ feedback ($v_-$):
$$H(jw)=\frac{A}{1+AB} \rightarrow_\infty \frac{1}{B} = 1+\frac{Z_f}{Z_g}$$

Posetiv feedback ($v_+$):
$$H(jw)=\frac{A}{1-AB}$$
For en ideel op-amp (A→∞) betyder 1−Aβ typisk ustabilitet (da Aβ≫1 ved lave frekvenser). Kredsløbet vil derfor saturere og opføre sig som en komparator/Schmitt-trigger — ikke som en lineær forstærker — og en meningsfuld lineær H(jω) findes ikke i praksis.

$\textbf{Samling af div. formler:}$
I den lukkede sløjfe indgår den i standardformlen: $\\$
$$H(jw)=\frac{A(jw)}{1+A(jw)\cdot B(jw)}$$
For en “ideel” op-amp tænker man $A\rightarrow\infty$, og så bliver $H≈\frac{1}{B}$.

Beedbackfaktoren $B(jw): \\$
$$B(jw) = \frac{Z_g}{Z_f+Z_g}$$

Åbensløjfe-forstærkning $A(jω): \\$
$$Y(jw)=A(jw)(V_+(jw)-V_-(jw))$$
Ofte modelleres A(jω) som et dominerende 1.-ordens led (dominerende pol):
$$A(jw)≈\frac{A_0}{1+j\frac{w}{w_p}}$$
Hvor $A_0$ er DC-forstærkningen (fx. $10^5 = 100 dB$), $ω_p$ er den dominerende pol (lille frekvens). $\\$

En nyttig størrelse er unity-gain båndbredden $ω_t$ (eller GBW i Hz): $\\$
$w_t = A_0w_p \Rightarrow f_t=A_0f_p$

> Opgave A.

$\textbf{Opgaven: } \\$
Et signal med Fourier transformationsparret $x(t) \leftrightarrow X(w)$ samples med tidsintervallet $\textbf{T}$ mellem samples.$\\$
Det samplede signal $\hat{x}(t) = x(t) \sum^\infty_{n=-\infty} \delta(t-nT)$ har et frekvensspektrum beskrevet ved:
$$\hat{X}(w)=\frac{1}{T}\sum^\infty_{n=-\infty} X(w-nw_s)$$
Det følger at frekvensspektret for det samplede signal er diskret med en afstand $w$, mellem frekvenskomposanterne på frekvensaksen.

$\textbf{Model for sampling i tid:} \\$
$$\hat{x}(t)=x(t)\sum^\infty_{n=-\infty}\delta(t-nT)=\sum^\infty_{n=-\infty}x(nT)\delta(t-nT)$$

$\textbf{Fouriertransform:} \\$
FT af kronefunktionen (Dirac comb) er også en comb:
$$\mathcal{F}\{\sum_n \delta(t-nT)\}=\frac{2\pi}{T}\sum_k\delta(w-kw_s), \qquad w_s=\frac{2\pi}{T}.$$

$\textbf{Produkt i tid ⇒ konvolution i frekvens}: \\$
$$\hat{X}(w)=\frac{1}{2\pi}X(w)*[\frac{a\pi}{T}\sum_k\delta(w-kw_s)]=\frac{1}{T}\sum^\infty_{k=-\infty}X(w-kw_s).$$

Konklusjon: Formlen i udsagnet er altså korrekt.

$\textbf{Hvad betyder det?} \\$
$\hat{X}(w)$ er en sum af forskudte kopier af det kontinuerte spektrum $X(ω)$. Det er derfor periodisk med periode $ω_s$ – ikke diskrete spektrallinjer. (Kun hvis $X(ω)$ selv er impuls-tog, bliver resultatet et impuls-tog.)

> Opgave B. 

$\textbf{Opgaven:} \\$
Et signal $x(t)$ indeholder en enkel frekvenskomponent ved 4Hz. Hvis signalet samples med $6Hz$, vil der i frekvensspektret for det samplede signal optræde en frekvenskomponent ved $2Hz$.


$\textbf{Forklaring:} \\$
Signal: én frekvenskomponent ved $𝑓_{signal}=4Hz$ (En enkel frekvenskomponent betyder bare, at signalet består af én ren sinus- eller cosinusbølge ved én bestemt frekvens.) $\\$
Samplingsfrekvens: $𝑓_𝑠=6Hz \\$ 
Påstand: Efter sampling vil der være en frekvenskomponent ved 2 Hz. $\\$

(Et frekvensspektrum er en beskrivelse af, hvor meget af et signals energi eller amplitude der findes ved hver frekvens.) $\\$
(En frekvenskomponent er én bestemt “byggeblok” i et signal — en ren sinus- eller cosinusbølge med én bestemt frekvens, amplitude og fase. - $\\$
Ifølge Fourier-teorien kan ethvert signal (næsten) beskrives som en sum af mange rene sinusbølger med forskellige frekvenser. Hver af disse sinusbølger er en frekvenskomponent.)

$\textbf{Aliasingformlen:} \\$
Når et signal med frekvens f samples med $f_s$, så “foldes” højere frekvenser ind i Nyquist-intervallet:
$$f_{alias} = |f-kf_s|$$
Hvor: $\\$
$f_{alias}$ Er den aliasede frekvens — den frekvens, vi “ser” i det samplede signal, efter højere frekvenskomponenter er foldet ind i Nyquist-intervallet $[0, \frac{f_s}{2}]. \\$
Dette er den “fejlagtige” lavere frekvens, som høje frekvenser bliver forvekslet med, når vi sampler for langsomt. $\\$
$f_s$ Samplingsfrekvensen — hvor mange gange per sekund vi tager prøver af signalet, måles i Hz. $\\$
$k$ Et heltal (0,1,2,...) der angiver, hvor mange hele samplingsfrekvenser $f_s$ vi flytter signalets frekvens med for at finde den foldede (aliasede) frekvens.

$\textbf{Beregning: } \\$
Her:
* $f=4Hz$
* $f_s=6Hz$
* Nyquist-frekvens: $f_N = \frac{f_s}{2} = 3 Hz $

Da $4 \text{ (frekvenskomponenten) } >𝑓_𝑁​$, sker aliasing. Vi finder k:
$$k = round(\frac{f}{f_s})=round(\frac{4}{6})=1$$
$$f_{alias} = |f-kf_s| \rightarrow f_{alias} = |4-1\cdot6|=|4-6|=2Hz$$

$\textbf{Konklusion: } \\$
Udsagnet siger, at vi får en komponent ved 2 Hz, og beregningen viser præcis det samme.

> Opgave C. ADC og sampeling.

$\textbf{Opgaven:} \\$
$\textbf{Udsagn C.} \\$
Figur A viser relationen mellem det analoge signal (blå graf) og det digitale signal (røde trappetrinskurve) for $ADC_A$, der altid runder amplituden på den digitale sample nedad til nærmeste mulige heltal. $\\$
Figur B viser relationen mellem det analoge signal (blå graf) og det digitale signal (røde trappetrinskurve) for $ADC_B$, der runder amplituden på den digitale sample op eller ned til nærmeste mulige heltal.

Amplituden på kvantiseringsstøjen for $ADC_A$ har en uniform statistisk fordelingsfunktion i intervallet $[0, \Delta V_{LSB}]. \\$
Amplituden på kvantiseringsstøjen for $ADC_B$ har en uniform statistisk fordelingsfunktion i intervallet $[-\Delta V_{LSB}/2, \space \Delta V_{LSB}/2]. \\$
Det er muligt at reducere kvantiseringsstøjen ved at oversample med en faktor M og udregne en middelværdi af M samples. Udbyttet af oversampling er dog langt ringere for $ADC_A$ end for $ADC_B$

$\textbf{Hvad er en ADC:} \\$
En ADC står for Analog-to-Digital Converter — altså en analog-til-digital-konverter. $\\$
Den gør kort fortalt dette:
* Modtager et analogt signal – f.eks. en spænding, der varierer kontinuerligt over tid.
* Tager prøver af signalet på faste tidspunkter (sampling).
* Kvantisere værdien – oversætter den målte spænding til det nærmeste trin på en digital skala (måleenheden her er typisk en “LSB” – Least Significant Bit).
* Sender resultatet som et digitalt tal, så det kan behandles af en mikrocontroller, computer eller DSP.

$\textbf{Forskellen på } ADC_A \textbf{og } ADC_B \textbf{i opgaven} \\$
$ADC_𝐵$: (rounding): Fejlen er biasfri → oversampling kan fjerne næsten al kvantiseringsstøj. $\\$
$ADC_𝐴$: (flooring): Fejlen har bias → gennemsnit kan reducere varians, men biasen forbliver, så forbedringen er meget mindre.

$\textbf{Løsning:} \\$
$ADC_A$ (flooring) Amplituden på kvantiseringsstøjen for $ADC_A$ har en uniform statistisk fordelingsfunktion i intervallet $[0,ΔV_{LSB}]$.
* Uniform fordeling: alle fejlværdier indenfor intervallet er lige sandsynlige.
* Intervallet starter ved 0 og går til $ΔV_{LSB}$, fordi $ADC_A$ altid runder ned → fejlen er altid positiv (målingen er altid lavere end den sande værdi).
* Dette giver en positiv bias: gennemsnitsfejl = $\frac{ΔV_{LSB}}{2}​$.

$ADC_B$ (rounding) Amplituden på kvantiseringsstøjen for $ADC_B$ har en uniform statistisk fordelingsfunktion i intervallet: $[\frac{-\Delta V_{LSB}}{2}, \frac{\Delta V_{LSB}}{2}]$
* Her kan fejlen være både negativ og positiv, fordi $ADC_B$ runder til nærmeste trin.
* Bias = 0, fordi fejlene er symmetrisk fordelt omkring nul.

Det er muligt at reducere kvantiseringsstøjen ved at oversample med en faktor M og udregne en middelværdi af M samples.
* Ved oversampling (tage flere samples end nødvendigt og gennemsnitliggøre) falder variansen af støjen med $\frac{1}{M}
* Det kan oversættes til flere effektive bits via formlen:
$$\Delta N_{bits} = \frac{1}{2}log_2(M)$$

Hvorfor det virker dårligere for $ADC_A: \textit{ Udbyttet af oversampling er dog langt ringere for } ADC_A \textit{ end for } ADC_B $
* For $ADC_B$: Støjen er biasfri → gennemsnit kan reducere både positive og negative fejl → støjen kan næsten elimineres.
* For $ADC_A$: Støjen har en fast positiv bias (Δ/2), som ikke reduceres ved gennemsnit. → Du kan kun reducere den tilfældige del af støjen, ikke det systematiske offset.

$\textbf{Opløsning og trinbredde: (LSB)} \\$
$$\Delta V_{LSB} = \frac{V_{ref-max} - V_{ref-min}}{2^N}$$
Hvor: $\\$
$N$ = antal bit. $\\$
$V_{ref}$ = referancespænding (eller måleområde)

$\textbf{Kvantifisering:} \\$
Den gennerelle formelen er:
$$Q(x)=\Delta V_{LSB} \cdot round(\frac{x}{\Delta V_{LSB}})$$
Her bliver verdien rundet op eller ned alt etter hvad som er tættest på.

Men for denne opgaven runder vi istede ned for $Q_A$
$Q_A(x)=\Delta V_{LSB} \cdot \lfloor\frac{x}{\Delta V_{LSB}}\rfloor \\$


$\textbf{Kvantiseringens fejl: } \\$
En ADC kan ikke måle helt præcist — den kvantiserer signalet til det nærmeste trin. $\\$
Kvantiseringens fejl er forskellen mellem den rigtige værdi og den kvantiserede værdi.

Den gennerelle formelen for fejl er:
$$e(x)=x-Q(x)$$
Hvor: $\\$
$x$ = den sande (analoge) værdi $\\$
$Q(x)$ = den kvantiserede værdi (altså trinværdien, ADC’en giver) $\\$
$e(x)$ = fejlen — positiv, negativ eller nul afhængigt af metode.

Bias (middelværdien af fejlen) er: $\\$
$$𝜇=𝐸[𝑒]$$

Varians: (spredningen af fejlen) er: $\\$
$$σ^2=E[(e−μ)^2]$$
$E[\cdot]$ betyder gennemsnit, enten over tid eller over en sandsynlighedsfordeling.

Ex. for rounding of flooring: $\\$
For B: $e_B=[-\frac{\Delta V_{LSB}}{2},\frac{\Delta V_{LSB}}{2}]$, $μ_B=\frac{a+b}{2}=\frac{-\frac{\Delta V_{LSB}}{2} + \frac{\Delta V_{LSB}}{2}}{2} = 0$, $ $σ^2_B=\frac{(b-a)^2}{12}$, $b-a=\Delta \Rightarrow σ^2_B=\frac{\Delta^2}{12}$

Og for A: $μ_A=\frac{0+\Delta}{2}=\frac{\Delta}{2}$, $σ^2_A=\frac{\Delta^2}{12}$

$\textbf{Signal-to-Quantization-Noise Ratio (SQNR):} \\$
Signal-to-Quantization-Noise Ratio (SQNR) er et mål for, hvor meget stærkere det nyttige signal er i forhold til kvantiseringsstøjen, når et analogt signal er blevet konverteret til digital form af en ADC.
$$SQNR=\frac{\text{signalets gennemsnitlige effekt}}{\text{signalets gennemsnitlige effekt}}$$
Eller i deciblel:
$$SQNR_{dB}=10log_{10}(\frac{P_{støj}}{P_{signal}})$$

Standardformel for en ideel N-bit ADC med sinusinput:
$$SQNR_{dB} \approx 6.02N+1.76$$
N = antal bits i ADC’en. $\\$
6.02 dB pr. bit → hver ekstra bit fordobler opløsningen og reducerer kvantiseringsstøjen med en faktor 4 i effekt. (6.02 kommer fra logaritmen af 4, når man regner i decibel.$\\$
1.76 dB er en konstant, der kommer af forholdet mellem sinus’ amplitudefordeling og uniform støj.

$\textbf{Oversampling-effekt} \\$
Oversampling-effekt handler om at forbedre opløsning og reducere støj i en ADC ved at tage flere målinger, end det strengt taget er nødvendigt, og så kombinere dem.

Antag at kvantiseringsfejl $e$ er tilfældig og uafhængig mellem samples.Hvis vi tager M samples og laver gennemsnittet:
$$\overline{x}=\frac{1}{M}\sum^M_{i=1} x_i$$
så bliver variansen:
$$σ^2_{avg}=\frac{σ^2}{M}$$
hvor $M$ er antal samples, der gennemsnittes. $\\$
Støjens rms-værdi falder med $\sqrt{M}$ 

$\textbf{Ekstra bits fra oversampling:} \\$
“Ekstra bits fra oversampling” betyder, at hvis du tager mange flere målinger (samples) end det minimum, der skal til for at dække signalets båndbredde, så kan du reducere kvantiseringsstøjen så meget, at det svarer til, at din ADC har fået flere bits opløsning – uden at du fysisk har ændret ADC’en.

Den teoretiske gevinst i effektiv opløsning er:
$$\Delta N_{bits} = \frac{1}{2}log_1(M)$$
M = hvor mange gange du oversampler og gennemsnitliggør. $\\$
Hver gang M ganges med 4, får du +1 ekstra bit.



> Opgave D: 

$\textbf{Opgaven:} \\$
Et analogt signal lavpasfiltreres af et 4. ordens lavpasfilter med en 3dB knækfrekvens på $100Hz$ og digitaliseres herefter af en $12-bit$ ADC. Hvis filteret skal dæmpe signalet så meget ved den halve samplingsfrekvens, at signalets amplitude ved denne frekvens er for lavt til at kunne registreres af ADC'en, skal samplingsfrekvensen være $1200Hz$.

$\textbf{Forklaring:} \\$
Filter: 4. ordens lavpas, $f_c=100 Hz$ (−3 dB ved knækfrekvensen (cutoff frequency) — punktet, hvor filteret begynder at dæmpe). $\\$
* I et Butterworth-lavpas (det mest almindelige) er signalets amplitude her faldet til $\frac{1}{2}$ af fuld styrke, det svarer til:
$$10log_{10}(\frac{1}{\sqrt{1}})=-3dB$$
* Derfor siger man, at filteret har −3 dB ved knækfrekvensen.
ADC: 12-bit → mindste registrerbare amplitude ≈ 1 LSB. $\\$
Krav: Ved $\frac{f_s}{2}$ (Nyquist-frekvensen) skal signalet være så dæmpet, at amplituden er lavere end 1 LSB, så ADC’en ikke registrerer det. $\\$
Påstand: Dette krav opfyldes, hvis $f_s=1200 Hz. \\$

$\textbf{Mindste registrerbare amplitude:} \\$
En 12-bit ADC har $2^12=4096$ trin. $\\$

Mindste trin:
$$LSB = \frac{V_{FS}}{4096}$$

Hvis vi normaliserer fuldskala til 1 V:
$$LSB = \frac{V_{FS}}{2^{12}} = \frac{1}{4096}V ≈ 2.44×10^{−4} \Rightarrow 20log_{10}(2.44×10^{−4}) = -72.24 dB$$
Så ét LSB svarer til ca. -72 dBFS.

$\textbf{Filterdæmpning ved } \frac{f_s}{2}: \\$
For et 4. ordens Butterworth-lavpas er amplitudeoverførslen:
$$|H(f)|= \frac{1}{\sqrt{1+(\frac{f}{f_c})^{2n}}}$$
med n=4

Dæmpning i dB:
$$A(f)=−10log_{10}(1+(f/f_c)^8)$$

$\textbf{Kravet til dæmpning:} \\$
Vi vil have:
$$|H(\frac{f_s}{2})| = \frac{1}{4096}$$
I dB:
$$A(\frac{f_s}{2})<-71dB$$

$\textbf{Find } f_s: \\$
Kravet:
$$\frac{1}{\sqrt{1+(\frac{f_s}{2f_c})^8}} = \frac{1}{4096}$$

Kvadrer begge sider:
$$\frac{1}{\sqrt{1+(\frac{f_s}{2f_c})^8}} = \frac{1}{4096} \Rightarrow \frac{1}{1+(\frac{f_s}{2f_c})^8} = \frac{1}{4096^2} \Rightarrow 1+(\frac{f_s}{2f_c})^8 = 4096^2 \Rightarrow \frac{f_s}{2f_c}^8 = 1.6777 \cdot 10^7$$

8.-roden:
$$\frac{f_s}{2f_c}≈(1.6777\cdot 10^7)^{\frac{1}{8}}≈7.62$$
$$f_s≈2f_c \cdot 7.62≈2 \cdot 100 \cdot 7.62≈1524 Hz$$

Påstanden siger 1200 Hz, beregningen giver ca. 1524 Hz som minimum. → Påstanden er falsk, fordi den foreslåede samplingsfrekvens er for lav til at opnå -72 dB dæmpning ved Nyquist. Den er falsk.
